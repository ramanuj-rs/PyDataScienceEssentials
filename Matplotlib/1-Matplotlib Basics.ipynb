## Basic Plot using Matplotlib


```python
plt.plot(x, y, 'b') # 'b' is the color parameter to the function here 'b' represents blue
plt.xlabel('X Axis Name')
plt.ylabel('Y Axis Name')
plt.title('X vs Y Plot')
plt.show()
```

## Generating multiple plots on a shared canvas.

```python
# subplot method prototype : plt.subplot(nnumber of rows, nubmer of columns, plot number)
plt.subplot(1,3,1)
plt.plot(x, y, 'g--') # Color along with line formatting
plt.subplot(1,3,2)
plt.plot(y, x, 'r');
plt.subplot(1,3,3)
plt.plot(y, x, '*-');
```

## Object Oriented Method

The fundamental concept behind adopting the more structured Object-Oriented method is to instantiate figure objects and subsequently invoke methods or access attributes from that object. This methodology is more convenient when working with a canvas containing several plots.

```python
fig = plt.figure() #Empty canvas to start

# Addition of set of axes to the canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height

# Plot on the canvas
axes.plot(x, y, 'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

## Nested Plot (Plot inside another plot)

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Aes for the outer canvas
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #Axes for the inner canvas

# Outer Plot
axes1.plot(x, y, 'r')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Outer Plot Title')

# Inner Plot
axes2.plot(y, x, 'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Inner Plot Title');
```

## Subplots using Object Oriented Method


```python
# Empty canvas
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
for ax in axes: #Iterate through different axex
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the plot
fig
```

### Maintain figure size, aspect ratio and DPI

Matplotlib enables the specification of the aspect ratio, DPI (dots per inch), and figure size during the creation of the Figure object. The figsize and dpi keyword arguments can be employed for this purpose.

```python
fig = plt.figure(figsize=(4,4), dpi=200)
```

```python
fig, axes = plt.subplots(figsize=(4,3))

axes.plot(x, y, 'r')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title');
```

## Saving figures
Matplotlib plots can be saved in different formats including PNG, JPG, EPS, SVG, PGF and PDF etc. To store a figure in a file, the savefig method within the Figure class can be utilized:







```python
fig.savefig("filename.png")
```

While saving using 'savefig()' method Specific target DPI can be passed as a parameter:

```python
fig.savefig("filename.png", dpi=200)
```

## Legends

Having covered the fundamental aspects of crafting a figure canvas and appending axes instances to the canvas, let's explore the process of embellishing a figure with legends.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x square")
ax.plot(x, x**3, label="x cube")
ax.legend()
```

## Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'r*-') #Red color with *- like line
ax.plot(x, x**3, 'b--')
```

### Setting colors with specific color hex code value

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="red", alpha=0.5) #alpha parameter for line transparency 0.5 represnt half tranparent
ax.plot(x, x+2, color="#8B008B")
ax.plot(x, x+3, color="#FF8C00")
```

### Linewidth

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```
