# Different NumPy Operations

## Arithmetic Operations

```python
import numpy as np
arr = np.arange(0,10)
```

```python
arr + arr
```

```python
arr * arr
```

```python
arr - arr
```

```python
# Caution regarding division by zero, but it won't cause an error! Instead, it is substituted with "nan" (Not a Number).
arr/arr
```

```python
1/arr
```

```python
arr**3
```

## Some Trigonometric Array Functions

```python
#Taking Square Roots
np.sqrt(arr)
```

```python
#Calcualting exponential (e^)
np.exp(arr)
```

```python
np.max(arr) #same as arr.max()
```

```python
np.sin(arr)
```

```python
np.log(arr)
```
