# NumPy Indexing and Selection


NumPy indexing and selection refer to the techniques and methods used to access and manipulate elements within NumPy arrays. Understanding these concepts is crucial for effective data manipulation and analysis using NumPy in Python.

```python
import numpy as np
```

```python
arr = np.arange(0,11)
```

```python
arr
```

## Bracket Indexing and Selection
Bracket indexing and selection in NumPy involve using square brackets [] to access and manipulate elements within NumPy arrays. This method is fundamental for data manipulation and analysis using NumPy in Python.

```python
#Get a value at an index
arr[8]
```

```python
#Get values in a range
arr[1:5]
```

```python
#Get values in a range
arr[0:5]
```

## Broadcasting

Broadcasting is a powerful concept in NumPy that enables the manipulation of arrays with different shapes during arithmetic operations. It allows arrays of different shapes to be combined seamlessly, eliminating the need for explicit replication of values and promoting more concise and efficient code.

```python
#Setting a value with index range (Broadcasting)
arr[0:5]=100

#Show
arr
```

```python
# Reset array, we'll see why I had to reset in  a moment
arr = np.arange(0,11)

#Show
arr
```

```python
#Important notes on Slices
slice_of_arr = arr[0:6]

#Show slice
slice_of_arr
```

```python
#Change Slice
slice_of_arr[:]=99

#Show Slice again
slice_of_arr
```

```python
#To get a copy,  it needs to be done explicitly
arr_copy = arr.copy()

arr_copy
```

## Indexing a 2D array (matrices)

Indexing a 2D array, commonly referred to as matrices, in NumPy involves accessing and manipulating elements within a two-dimensional structure.Accessing a single element in a 2D array involves specifying the row and column indices.

```python
arr_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))

arr_2d
```

```python
#Indexing row
arr_2d[1]
```

```python
# Getting an element value based on row and column indices
arr_2d[1][0]
```

```python
# Another way to use row and column indices
arr_2d[1,0]
```

```python
# 2D array slicing

#Shape (2,2) from top right corner
arr_2d[:2,1:]
```

```python
#Shape bottom row
arr_2d[2]
```

```python
#Another way to Shape bottom row
arr_2d[2,:]
```

### Fancy Indexing

Fancy indexing in NumPy refers to the technique of accessing and manipulating elements in an array using arrays of indices or boolean arrays. It provides a flexible and powerful way to make complex selections and modifications in arrays. Fancy indexing is particularly useful when you need to extract or modify specific elements based on non-trivial conditions or specific positions.

```python
#Set up matrix
arr2d = np.zeros((10,10))
```

```python
#Length of array
arr_length = arr2d.shape[1]
```

```python
#Set up array

for i in range(arr_length):
    arr2d[i] = i

arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
#Allows in any order
arr2d[[6,4,2,7]]
```

## Selection

Selection in NumPy involves choosing and extracting specific elements or subsets from an array. It encompasses various techniques, including basic indexing, slicing, boolean indexing, and fancy indexing, providing a versatile set of tools for data manipulation.


```python
arr = np.arange(1,11)
arr
```

```python
arr > 4
```

```python
bool_arr = arr>4
```

```python
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[arr>2]
```

```python
x = 2
arr[arr>x]
```
