# Operations

Operations in pandas refer to various actions and calculations that can be performed on Series and DataFrames, enabling users to manipulate and analyze data efficiently. Pandas provides a wide range of operations, including arithmetic, statistical, logical, and custom operations.

```python
import pandas as pd
df = pd.DataFrame({'Roll':[1,2,3,4],'Marks':[80,95,80,85],'Name':['Ramanuj','Shayak','Krishanu','Saptarsi']})
df.head()
```

### Unique Values

```python
df['Marks'].unique()
```

```python
df['Marks'].nunique()
```

```python
df['Marks'].value_counts()
```

### Selecting Data

```python
newdf = df[(df['Roll']>1) & (df['Marks']==80)]
```

```python
newdf
```

### Applying Functions

```python
def times2(x):
    return x*2
```

```python
df['Roll'].apply(times2)
```

```python
df['Name'].apply(len)
```

```python
df['Roll'].sum()
```
