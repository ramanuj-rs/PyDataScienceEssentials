# Groupby

The groupby function in pandas is a powerful tool for grouping and aggregating data based on specified criteria. It enables users to split a DataFrame into groups based on one or more columns and then apply a function to each group independently. This is particularly useful for tasks such as data segmentation, summary statistics, and group-wise operations.

```python
import pandas as pd

data = {'Role':['Prof','Prof','Sde','Sde','Tester','Sde'],
       'Person':['Ramanuj','Souvik','Shayak','Hunnan','Sandeep','Krishanu'],
       'Salary':[10,8,20,11,17,15]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Role')
```

```python
by_role = df.groupby("Role")
```

```python
by_role.mean()
```

```python
df.groupby('Role').mean()
```

```python
by_role.std()
```

```python
by_role.min()
```

```python
by_role.max()
```

```python
by_role.count()
```

```python
by_role.describe()
```

```python
by_role.describe().transpose()
```

```python
by_role.describe().transpose()['Sde']
```
