# Missing Data

Handling missing data is a crucial aspect of data cleaning and preprocessing in data analysis. In pandas, missing data is often represented as NaN (Not a Number). The library provides several tools and methods to effectively manage missing data in DataFrames.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,np.nan,np.nan],
                  'B':[4,5,np.nan],
                  'C':[7,8,9]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```
