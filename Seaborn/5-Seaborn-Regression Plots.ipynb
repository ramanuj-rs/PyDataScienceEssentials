# Regression Plots

Regression plots in Seaborn are dedicated to visualizing the relationship between two numerical variables and fitting a regression model to the data. These plots provide a clear representation of the trend and help assess the strength and direction of the linear relationship between the variables. Seaborn offers various regression plot functions, such as lmplot and regplot, allowing users to customize the appearance and style of the plots to suit their analysis needs. Regression plots are valuable for understanding the correlation and predicting the outcome of one variable based on another in a dataset.

```python
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',palette='coolwarm')
```

## Using a Grid

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='day',hue='sex',palette='coolwarm')
```
