# Categorical Data Plots

There are various function in seaborn to plot categorical data. These are as follows :

* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head()
```

## barplot

The barplot is a versatile plot that permits the aggregation of categorical data using a specified function, with the mean being the default.

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

### countplot

Countplot is identical to a barplot, except the estimator explicitly counts the number of occurrences. Hence, we only provide the x value.

```python
sns.countplot(x='sex',data=df)
```

## boxplot

Boxplots are employed to illustrate the distribution of categorical data. A box plot, also known as a box-and-whisker plot, represents the distribution of quantitative data, facilitating comparisons between variables or across levels of a categorical variable. The box delineates the quartiles of the dataset, while the whiskers extend to display the remainder of the distribution, excluding points identified as "outliers" based on a method that depends on the inter-quartile range.

```python
sns.boxplot(x="sex", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="sex", y="total_bill", hue="smoker",data=df, palette="coolwarm")
```

### violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="sex", y="total_bill", data=df,palette='rainbow')
```

```python
sns.violinplot(x="sex", y="total_bill", data=df,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="sex", y="total_bill", data=df,hue='sex',split=True,palette='Set1')
```

## stripplot and swarmplot


The stripplot generates a scatterplot when one variable is categorical. It can be presented independently, but it also serves as a useful addition to a box or violin plot when there is a need to display all observations alongside a representation of the underlying distribution.

The swarmplot resembles stripplot(), but the points are shifted (solely along the categorical axis) to prevent overlap. This results in a more accurate representation of the value distribution, although it may not be as efficient for large numbers of observations, both in terms of displaying all points and the computational effort required to organize them.

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
sns.stripplot(x="day", y="total_bill", data=df,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=df,jitter=True,hue='sex',palette='Set2')
```

```python
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=df, palette="Set2")
```

### Combinied Categorical Plots

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
sns.swarmplot(x="total_bill", y="day", data=df,color='black',size=3)
```
