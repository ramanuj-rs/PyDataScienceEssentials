**Variables**: Reserved momory location to store values.
The eual sign (=) is used to assign values to variables. the operator to the left of the = operator is name of the variable and the operand to the next to the right of the = operator is the value stored in the variable.

```python
a=100  #An integer assignment
b=100.30  #An floating point
c="Python"   #A string
print(a)
print(b)
print(c)
```

Assigning a single value to several variables

```python
a=b=c=2
print(a)
print(b)
print(c)
```

Assigning multiple values to multiple variables

```python
a,b,c=1,2,"Hello"
print(a)
print(b)
print(c)
```

## Operators

```python
# 1) Arithmetic operator
a,b=11,5
print("sum =", a+b)
print("Difference", a-b)
print("Product", a*b)
print("Quotient", int(a/b))
print("Quotient", a/b)
print("Remainder", a%b)
```

```python
# 2) Comparison operator
a,b=11,5
print("a==b ", (a==b))
print("a!=b", (a!=b))
print("a>b", (a>b))
print("a<b", (a<b))
print("a>=b", a>=b)
print("a<=b", a<=b)
```

```python
# 3) Logical operator
a,b,c,d=10,20,30,40
print((a>b) and(c<d))
print((a>b) or (c<d))
print(not (a>b))
```
